```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df_2021 = pd.read_csv('verkehrsmittelwahl2021.csv', delimiter=';')
```

```python
# Convert relevant columns to numeric types
numeric_columns = ['BICYCLE', 'BIKESHARING', 'BY_FOOT', 'CAR', 'CARSHARING', 'MOTORBIKE', 'PUBLIC_TRANSPORT']
df_2021[numeric_columns] = df_2021[numeric_columns].apply(lambda x: x.str.replace(',', '.')).apply(pd.to_numeric)

# Plot the development over the years
plt.figure(figsize=(10, 6))

for column in numeric_columns:
    plt.plot(df_2021['YEAR'], df_2021[column], label=column)

plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Development Over the Years')
plt.legend()
plt.show()
```
